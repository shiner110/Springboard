## Part 1 ‚Äê Exploratory data analysis

The attached logins.json file contains (simulated) timestamps of user logins in a particular
geographic location. Aggregate these login counts based on 15minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the
underlying patterns of the demand. Please report/illustrate important features of the demand,
such as daily cycles. If there are data quality issues, please report them.

```python
# Import relevent libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Upload login information into a dataframe
df = pd.read_json('logins.json')
```

```python
# Explore the data
df.info()
```

```python
# First few rows
df.head()
```

```python
# Time series, convert index to datetime
df = df.set_index('login_time')
df.index = pd.to_datetime(df.index)
```

```python
# Add column of 1's as a 'count' column to aggregate over
df['count']=1
```

```python
# Look at the dataframe
df.head()
```

```python
# Visualize counts for 15 minute intervals
fifteen = df.resample('15Min').sum()
ax = fifteen.plot(figsize=(20,8), label='15 Min', legend=False,  color = '#F1948A')
ax.set_ylabel('Count')
ax.set_title('Time series of login counts every 15 minutes');
```

```python
# Check for null values
fifteen.isna().any()
```

```python
# Resample with a daily and weekly rolling averages

# Daily rolling average in BLUE
ax = fifteen.resample('d').mean().plot(figsize=(20,8), color='blue', linewidth=6, legend = False)

# Weekly rolling average in RED
fifteen.resample('w').mean().plot(ax = ax, color = 'red', linewidth = 4, legend = False)

# Original data in PINK
fifteen.plot(ax=ax, alpha=0.5, color='#F1948A', legend = False)

ax.set_title('Daily and Weekly rolling averages')
ax.set_ylabel('Counts')

plt.show()
```

Taking average daily values, there is a weekly cycle apparent in the data. 

On the other hand the data averaged over a week is relatively constant, with a gradual increase in spring. 

```python
# Zoom in on March and visualize 6 hour rolling average to get a better understanding of daily usage

# 6 hour rolling average BLUE
ax = fifteen.loc['1970-03'].resample('6h').mean().plot(figsize=(20,8), linewidth=4, legend = False)

# Original data ORANGE
fifteen.loc['1970-03'].plot(ax = ax, alpha = 0.6, legend = False)
ax.set_title('6 Hour rolling average for February and March')
ax.set_ylabel('Counts')

plt.show()
```

The largest spikes happen over the weekend, when there is a double peak on Friday and Saturday followed by a lower average on Sundays. Monday, Tuesday and Wednesday see the least amount of logins with a gradual increase on Thursdays. 

```python
# 2 hour resample
ax = fifteen.loc['1970-02-02':'1970-02-05'].resample('2h').mean().plot(figsize=(20,8), linewidth=4)
fifteen.loc['1970-02-02':'1970-02-05'].plot(ax=ax, alpha=0.6);
```

Most logins happen around noon. There seems to be a second, smaller peak usage around midnight!
